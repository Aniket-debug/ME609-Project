```python
class bracket_operator_penalty:
    def __init__(self, d, f, g, h, l, u, j, k):
        self.d = d  # Number of variables
        self.f = f  # Objective Function
        self.g = g  # Nonlinear Constraints
        self.h = h  # linear Constrants
        self.l = l  # lower limit of x
        self.u = u  # upper limit of x
        self.j = j  # number of non linear constraints
        self.k = k  # number of linear constraints
        self.it_vs_x = []
        self.feval = []

    def plot_it_vs_feval(self):
        title = "Iteration vs Function Evaluations"
        xlabel = "Iteration"
        ylabel = "Function Evaluations"
        history = np.array(self.feval, dtype=object)

        iterations = range(1, len(self.feval) + 1)

        plt.figure(figsize=(6, 4))
        plt.plot(iterations, self.feval, marker="o", linestyle="-", color="b")
        plt.title(title)
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        plt.grid(True)
        plt.show()

    def plot_it_vs_x(self):
        title = "Iteration vs Objective Function Value"
        xlabel = "Iteration"
        ylabel = "Objective Function Value"
        history = np.array(self.it_vs_x, dtype=object)

        iterations = [item[0] for item in history]
        x_vectors = np.array([item[1] for item in history])

        function_values = np.array([self.f(x) for x in x_vectors])

        plt.figure(figsize=(6, 4))
        plt.plot(iterations, function_values, marker="o", linestyle="-", color="b")
        plt.title(title)
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        plt.grid(True)
        plt.show()

    def countour_plot(self):

        x1 = np.linspace(self.l[0], self.u[0], 500)
        x2 = np.linspace(self.l[1], self.u[1], 500)
        X1, X2 = np.meshgrid(x1, x2)

        Z = np.array(
            [self.f(np.array([x1, x2])) for x1, x2 in zip(X1.ravel(), X2.ravel())]
        ).reshape(X1.shape)

        G = np.array(
            [self.g(np.array([x1, x2])) for x1, x2 in zip(X1.ravel(), X2.ravel())]
        )
        G = G.reshape(X1.shape + (len(G[0]),))

        H = np.array(
            [self.h(np.array([x1, x2])) for x1, x2 in zip(X1.ravel(), X2.ravel())]
        )
        H = H.reshape(X1.shape + (len(H[0]),))

        feasible_region = np.all(G >= 0, axis=2) & np.all(np.abs(H) <= 0.01, axis=2)

        plt.figure(figsize=(8, 8))
        contour = plt.contour(
            X1,
            X2,
            Z,
            cmap="viridis",
        )
        plt.clabel(contour, inline=True, fontsize=8)
        plt.colorbar(contour, label="Objective function value")

        plt.contourf(X1, X2, feasible_region, levels=[0.5, 1], colors="red", alpha=0.5)
        plt.plot([], [], color="red", label="Feasible Region")

        history = self.it_vs_x
        trajectory = np.array([item[1] for item in history])
        for i in range(len(trajectory) - 1):
            plt.annotate(
                "",
                xy=trajectory[i + 1],
                xytext=trajectory[i],
                arrowprops=dict(arrowstyle="->", color="black", lw=1.5),
            )

        starting_point = trajectory[0]
        plt.scatter(
            starting_point[0],
            starting_point[1],
            color="blue",
            s=10,
            label="Starting Point",
        )

        min_x1, max_x1 = np.min(trajectory[:, 0]), np.max(trajectory[:, 0])
        min_y1, max_y1 = np.min(trajectory[:, 1]), np.max(trajectory[:, 1])

        padding_x = 0.05 * (max_x1 - min_x1)
        padding_y = 0.05 * (max_y1 - min_y1)
        plt.xlim(self.l[0] - padding_x, self.u[0] + padding_x)
        plt.ylim(self.l[1] - padding_y, self.u[1] + padding_y)

        plt.xlabel("$x_1$")
        plt.ylabel("$x_2$")
        plt.title("Contour plot with feasible region and trajectory")
        plt.legend()
        plt.grid(True)
        plt.show()

    def __call__(self, c=1.2, mom=False):
        eps1 = 10**-6
        eps2 = 10**-6
        r = 0.1
        # c = 1.2
        M = 19 / math.log10(c)
        k = 1
        sigma = np.zeros(self.j)
        tau = np.zeros(self.k)

        x = np.random.uniform(self.l, self.u)
        initial_guess = x
        self.it_vs_x.append((k, x))
        # print(f"initial guess {x}")

        if mom:
            M = 100

        while k <= M:

            penalty_fun = (
                lambda x, R=r: self.f(x)
                + R
                * np.sum(
                    np.square(np.minimum(self.g(x) + sigma, np.zeros(self.j)))
                    - np.square(sigma)
                )
                + R * np.sum(np.square(self.h(x) + tau) - np.square(tau))
            )

            optimizer = Multi_variable_optimizer(
                penalty_fun, 1000, self.l, self.u, self.d, eps1
            )

            x_next = optimizer.Conjugate_Gradient_Method(x)
            pf_x_next = penalty_fun(x_next)
            self.it_vs_x.append((k, x_next))

            if k > 1:
                # print(
                #     f"#it: {k} | R: {float(r):.2e} | X: {[f'{xi:.2e}' for xi in x_next]} | error: {abs(pf_x_next - pf_x):.2e}"
                # )
                if abs(pf_x_next - pf_x) <= eps2:
                    return x_next, k, initial_guess
            else:
                # print(
                #     f"#it: {k} | R: {float(r):.2e} | X: {[f'{xi:.2e}' for xi in x_next]} | error: None"
                # )
                pass

            if not mom:
                r *= c
            else:
                sigma = np.minimum(self.g(x_next) + sigma, np.zeros(self.j))
                tau = self.h(x) + tau
            x = x_next
            pf_x = pf_x_next
            self.feval.append(1 + optimizer.fun_evals)
            k += 1

        return x, k - 1, initial_guess
```

```python
def problem(value):
    d1 = 2

    l1 = np.array([13, 0])
    u1 = np.array([20, 4])

    f1 = lambda x: (x[0] - 10) ** 3 + (x[1] - 20) ** 3

    def g1(x):
        g11 = lambda x: (x[0] - 5) ** 2 + (x[1] - 5) ** 2 - 100

        g12 = lambda x: -1 * ((x[0] - 6) ** 2 + (x[1] - 5) ** 2 - 82.81)

        g1f = np.array([g11(x), g12(x)])

        return g1f

    def h1(x):
        return [
            0,
        ]  # As there is no linear constraints in problem1

    given_ox1 = [14.09500000000000064, 0.8429607892154795668]

    j1, k1 = 2, 0

    d2 = 2

    f2 = (
        lambda x: -1
        * (math.sin(2 * math.pi * x[0]) ** 3 * math.sin(2 * math.pi * x[1]))
        / (x[0] ** 3 * (x[0] + x[1]))
    )

    def g2(x):
        g21 = lambda x: -1 * (x[0] ** 2 - x[1] + 1)

        g22 = lambda x: -1 * (1 - x[0] + (x[1] - 4) ** 2)

        g2f = np.array([g21(x), g22(x)])

        return g2f

    def h2(x):
        return [
            0,
        ]  # As there is no linear constraints in problem2

    l2 = np.array([0.0001, 0.0001])
    u2 = np.array([10, 10])
    given_ox2 = [1.227, 4.245]
    j2, k2 = 2, 0

    match value:
        case 1:
            return d1, f1, g1, h1, l1, u1, given_ox1, j1, k1
        case 2:
            return d2, f2, g2, h2, l2, u2, given_ox2, j2, k2
```

```python
def main(problem_id, c=10):
    d, f, g, h, l, u, given_ox, j, k = problem(problem_id)

    optimizer = bracket_operator_penalty(d, f, g, h, l, u, j, k)

    optimal_x, iterations, initial_guess = optimizer(mom=False)
    min_f = f(optimal_x)
    total_fun_evals = sum(optimizer.feval)

    print("-" * 100)
    print(f"range of x {l} to {u}")
    print("Number of Iterations", iterations)
    print("Optimal X ", optimal_x)
    print("given Optimal X", given_ox)
    print(f"Objective Function Value at Optimal X {min_f}")
    print(f"Objective Function Value at given Optimal X {f(given_ox)}")

    feasiblity = np.all(np.array(g(optimal_x)) >= 0) and np.all(
        np.abs(np.array(h(optimal_x))) <= 0.01
    )

    feasiblity_given_ox = np.all(np.array(g(given_ox)) >= 0) and np.all(
        np.abs(np.array(h(given_ox))) <= 0.01
    )

    print(f"Optimal X is Feasible? {feasiblity}")
    print(f"Gvien Optimal X is Feasible? {feasiblity_given_ox}")

    optimizer.plot_it_vs_x()
    if d == 2:
        optimizer.countour_plot()
    optimizer.plot_it_vs_feval()

    return (
        optimal_x,
        iterations,
        feasiblity,
        min_f,
        initial_guess,
        l,
        u,
        total_fun_evals,
    )


# 1. 14.09500000000000064, 0.8429607892154795668 --> -6961.81387558015 ok
# 2. 1.227, 4.245 --> 0.09581940945335318 ok
```
